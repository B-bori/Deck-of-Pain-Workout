```python
import os
import random
import matplotlib.image as img
import matplotlib.pyplot as plt
import time
import winsound
```

```python
workout_type = {'hearts': input("Which workout will you do with hearts shape?:\t\t"),
                'diamonds': input("Which workout will you do with diamonds shape?:\t\t"),
                'clubs': input("Which workout will you do with clubs shape?:\t\t"),
                'spades': input("Which workout will you do with hearts shape?:\t\t")}

print()

reps = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9,
        'ten': 10, 'ace': input('how many reps will you do for ace cards?:\t\t'), 
        'jack': input('how many reps will you do for jack cards?:\t\t'), 
        'queen': input('how many reps will you do for queen cards?:\t\t'), 
        'king': input('how many reps will you do for king cards?:\t\t')}

print()

interval = int(input('alarm interval (seconds):\t\t'))

print()

print('='*100)
print()

playing_card_images = os.listdir('playing card images')

random.shuffle(playing_card_images)

for i in range(len(playing_card_images)):
    
    card = playing_card_images[i]
    image = img.imread('playing card images/{}'.format(card))
    
    num = card.split(' ')[1]
    shape = card.split(' ')[-1].split('.')[0]
    
    print('\033[1m' + '[{}]'.format(i+1) + ' Do {} for {} reps'.format(workout_type[shape], reps[num]))
    
    plt.imshow(image)
    plt.show()
    time.sleep(interval)
    winsound.Beep(500, 2000)
```
